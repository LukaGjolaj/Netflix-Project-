```python
#imports
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#reading data
df = pd.read_csv("Netflix TV Shows and Movies.csv")
df.head()
```

```python
#movies vs show cleaning 
movies_df = df[df['type'] == 'MOVIE']
shows_df = df[df['type'] == 'SHOW']

shows_df.head()
```

```python
#movies vs shows count
df['type'].value_counts().plot(kind='bar', color=['skyblue', 'lightgreen'])
plt.title('Movies vs TV Shows on Netflix')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
#means by type
mIMDBmean =movies_df["imdb_score"].mean()
sIMDBmean = shows_df["imdb_score"].mean()
totalIMDBmean = df["imdb_score"].mean()

labels = ['Movies', 'TV Shows','Total']
means = [mIMDBmean, sIMDBmean,totalIMDBmean]
plt.bar(labels, means, color=['skyblue', 'red', 'black'])
plt.ylabel('Average IMDb Score')
plt.title('Average IMDb Scores: Movies vs TV Shows')
plt.ylim(0, 10) 
```
